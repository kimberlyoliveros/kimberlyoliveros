Rock, Paper,Scissors

```python
import random

user_action = input("Enter a choice (rock, paper, scissors): ")
possible_actions = ["rock", "paper", "scissors"]
computer_action = random.choice(possible_actions)
print(f"\nYou chose {user_action}, computer chose {computer_action}.\n")

if user_action == computer_action:
    print(f"Both players selected {user_action}. Draw!")
elif user_action == "rock":
    if computer_action == "scissors":
        print("Rock smashes scissors! Ayy!")
    else:
        print("Paper covers rock! You suck.")
elif user_action == "paper":
    if computer_action == "rock":
        print("Paper beats the rock! Bravo!")
    else:
        print("Scissors beats paper! You're an L.")
elif user_action == "scissors":
    if computer_action == "paper":
        print("Scissors beats paper! Draw!")
    else:
        print("Rock smashes scissors! Sucks to suck haha")
```

Random Number generator 

```python
#random number generator
from random import randint #computer chooses random number

secret = randint(1,20) #the computer has a range between 1-50 to choose from

print("choose a number between 1 through 30")
a = int(input()) #I want user feedback

if a == secret: #If the user choose the correct input, they guessed correctly
    print("You are right!")
else:
    print("oops, not quite, try a lower number") #try again one last time
    b = int(input())
    if b == secret:
        print("You're uess it!!") #if they tried again and guess correctly, they got it.
    elif b < secret:
        print("nope") #guessed number is too low
        print("The number was", secret) #revealed
    elif b > secret:
        print("nope, number too low..") #too high
        print("The number was,", secret)
```
